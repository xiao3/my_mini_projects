# 湖北省疫情地图

```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Map
```

```python
# input data
df_input = pd.read_excel('../data/hubei_20200225.xlsx')
df_input
```

```python
districts = df_input['地区名'].tolist()
value = df_input['确诊人数'].tolist()
max_value = max(value)
min_value = min(value)
```

```python
# Define map info:
my_mark = "确诊地区"
my_prov = "湖北"
my_title = "新型冠状病毒感染肺炎湖北省疫情地图"
my_subtitle = "截至2020年2月25日  绘制：xxx"
my_output_path_1 = "../output/hubei-map-1.html"
my_output_path_2 = "../output/hubei-map-2.html"


c = (
        Map(init_opts=opts.InitOpts(width ='1000px', height='1000px'))
        .add(my_mark, [list(z) for z in zip(districts, value)], my_prov)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
        .set_global_opts(
            title_opts=opts.TitleOpts(title=my_title, subtitle=my_subtitle),
            visualmap_opts=opts.VisualMapOpts(min_=min_value, 
                                              max_=max_value,
                                              #is_piecewise=True,
                                             ),
        )
    )


c.render(path=my_output_path_1)
c.render_notebook()
```

```python
# Define map info:

my_pieces = [
      {"min": 10000, "label": ">10000", "color": "#8B1A1A"}, 
      {"min": 1001, "max": 10000, "label": "1001-10000", "color": "#CD3333"},
      {"min": 100, "max": 1000, "label": "101-1000", "color": "#FF6347"},
      {"max": 100, "label": "1-100", "color": "#FFA07A"}   
        ]



c = (
        Map(init_opts=opts.InitOpts(width ='1000px', height='1000px'))
        .add(my_mark, [list(z) for z in zip(districts, value)], my_prov)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
        .set_global_opts(
            title_opts=opts.TitleOpts(title=my_title, subtitle=my_subtitle),
            visualmap_opts=opts.VisualMapOpts(min_=min_value, 
                                              max_=max_value,
                                              is_piecewise=True,
                                              pieces=my_pieces,
                                             ),
        )
    )


c.render(path=my_output_path_2)
c.render_notebook()
```
