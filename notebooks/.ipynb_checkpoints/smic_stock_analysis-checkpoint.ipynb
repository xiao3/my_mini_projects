# SMIC Stock Analysis 

```python
import pandas as pd
from pandas_datareader import data
```

```python
smic = data.DataReader('0981.HK', start='2019', end='2020', data_source='yahoo')
```

```python
smic.head(10)
```

```python
len(smic['Close'])
```

```python
smic.tail(10)
```

```python
'''
定义函数
函数功能：计算股票涨跌幅=（现在股价-买入价格）/买入价格
输入参数：clo_column是收盘价这一列的数据
返回数据：涨跌幅
'''

def change(clo_column):
    #买入价格
    buyPrice=clo_column[0]
    #现在股价,序号是从0开始的，所以最后一条数据的序号是column的长度减1
    curPrice=clo_column[len(clo_column)-1]
    #累计涨跌幅
    priceChange=(curPrice-buyPrice)/buyPrice
    #判断股票是上涨，还是下跌
    if(priceChange>0):
        print('股票累计上涨=',priceChange)
    elif(priceChange<0):
        print('股票累计下跌=',priceChange)
    else:
        print('股票累计无变化=',priceChange)
    
    #返回数据
    return priceChange

#获取收盘价Close这一列的数据
closeCol=smic['Close']
#调用函数，获取涨跌幅
smicChange=change(closeCol)
```

```python
#导入绘图包
import matplotlib.pyplot as plt
#使用收盘价Close这一列数据绘制
fig= plt.figure(figsize=(10,4))
plt.plot(smic['Close'])
#做一个图片标题‘SMIC’
plt.title('SMIC')
#显示图片
plt.show()
```

```python
smic_2020 = data.DataReader('0981.HK', start='2020', end='2021', data_source='yahoo')
```

```python
#获取收盘价Close这一列的数据
closeCol=smic_2020['Close']
#调用函数，获取涨跌幅
smicChange=change(closeCol)
```

```python
#导入绘图包
import matplotlib.pyplot as plt
#使用收盘价Close这一列数据绘制
fig= plt.figure(figsize=(10,4))
plt.plot(smic_2020['Close'])
#做一个图片标题‘SMIC’
plt.title('SMIC')
#显示图片
plt.show()
```

```python
# 绘制图表
fig, ax = plt.subplots(figsize=(14,8)) # 绘图窗口大小
plt.title('2020 SMIC', fontsize=18)  # 标题
# 绘制参考线
((smic_2020['Adj Close']-smic_2020['Open'])).plot(grid=True, color='red', label='daily(close - open)').axhline(y=0, color='black', lw=2)
ax.legend(frameon=False)
plt.xlabel('1-12', fontsize=14)   # 横坐标
plt.ylabel('fluc', fontsize=14) #纵坐标
# 刻度标记大小
# axis 可选择参数 x,y,both 
plt.tick_params(axis='both',labelsize=13) 
plt.show() # 显示
```
