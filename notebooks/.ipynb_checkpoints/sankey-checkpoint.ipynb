# 桑基图

```python
import pandas as pd
from pyecharts.charts import Page, Sankey
from pyecharts import options as opts
```

```python
# 读取Excel文件
data = pd.read_excel('../data/airlines.xlsx', header=None)
```

```python
# 生成nodes
nodes = []
#nodes.append()
for i in data[0].unique():
    dic = {}
    dic['name'] = i
    nodes.append(dic)
 
for i in data[1].unique():
    dic = {}
    dic['name'] = i
    nodes.append(dic)
```

```python
nodes
```

```python
# 生成links
links = []
for i in data.values:
    dic = {}
    dic['source'] = i[0]
    dic['target'] = i[1]
    dic['value'] = i[2]
    links.append(dic)
```

```python
links
```

```python
# pyecharts 所有方法均支持链式调用。

my_title = "国际航班飞往中国大陆航班数量 桑基图"
my_output_path = "../output/airline-result.html"

c = (
        Sankey()
        .add(
"",
            nodes,
            links,
            linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color="source",type_="dotted"),
            label_opts=opts.LabelOpts(position="right"),
        )
        .set_global_opts(title_opts=opts.TitleOpts(title=my_title))
    )

# 输出html可视化结果
c.render(path=my_output_path)
c.render_notebook()
```
